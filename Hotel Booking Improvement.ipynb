# Data Cleaning and Modifying

```python
df.isnull()

# By using above method isnull() we can see that which are the places where there is any value or not
# If it is blank it will give us Ture and if it is not null or not NaN then it will give us False
```

```python
# Now I want to sum all the null values of each column and see how many blank spaces I have in which column
null = df.isnull()
count_null = null.sum()
print(count_null)
```

```python
# Now I will seperate the columns having null values and see which are those columns

df1 = pd.DataFrame([count_null])
# Here I have converted the previously found count_null items into a data frame

null_columns = []
for d in df1:
    if df1.loc[0,d] > 0:
        null_columns.append(d)
# Here I have removed the columns from the above data frame which has more than 0 null values and stored them inside a list

null_columns

```

```python
# Now we will see from the null value having columns which columns has how many nulls

for i in null_columns:
    print(f'{i} = {df1.loc[0,i]}')
```

```python
null_columns_values = []

for k in null_columns:
    null_columns_values.append(df1.loc[0,k])
# Here I have stored the sum of nulls with respected to null value holding columns

null_columns_values.sort()
# Here I have sorted the values in acending manner

null_columns_values = null_columns_values[::-1]
# Here I have sorted the values in descending manner

null_columns_values
```

```python
# Now we will also make the column names sorted as per the null_columns_values list
desc_null_columns = []

for a in range(len(null_columns)):
    for h in range(len(null_columns)):
        if null_columns_values[a] == df1.loc[0,(null_columns[h])]:
            desc_null_columns.append(null_columns[h])

desc_null_columns
```

```python
# Lets see the column names with sum of null values with respect to total rows

for m in desc_null_columns:
    print(f'{m} = {df1.loc[0,m]} / {len(df.index)}')
```

Here we can see that company column has most null values

```python
# Here we will find how much % is blank in the null value holding columns
for m in desc_null_columns:
    g = ((df1.loc[0,m]) / (len(df.index)))
    
    print(f'{m} = { g * 100} %')
```

In the above information we can see that comapny has excess null values about 94% and agent is also having 13% of null values.
Due to there excessive null value counts we can not use them to get the high probability insights from the data hence we will drop them.

```python
df.drop(['company','agent'], axis = 1)

# Here I have dropped the company and agent column but have not made it permanent
# I would now copy the data frame without having company and agent into a new data frame

df_no_null = df.drop(['company','agent'], axis = 1)
```

```python
# Lets see the new data frame
df_no_null
```

```python
# Now we can see there is a column having date and time named 'reservation_status_date'
# Lets see the data type of this column
df_no_null.loc[:,'reservation_status_date']
```

```python
''' As we can see here the data type is object of this column so now lets change the data type of this column 
from object to date and time'''

df_no_null.loc[:,'reservation_status_date'] = pd.to_datetime(df_no_null.loc[:,'reservation_status_date'])
# Now we can see the type has been changed from object to datetime64
```

```python
df_no_null.info()
```

```python
# Now we will drop all the remaining rows which are carrying null values
df_no_null = df_no_null.dropna(subset = ['children','country'])
```

```python
# # Here we will convert another column named children's data type from float to int64
df_no_null.loc[:,'children'] = df_no_null.loc[:,'children'].astype('int64')

df_no_null.info()
```

# Exploratory Data Analysis (EDA)

```python
# Here as we have better data set now we can perform some analysis on it and get insights
df_no_null.iloc[:,0:10].head()
df_no_null.iloc[:,10:20].head()
df_no_null.iloc[:,20:30].head()
df_no_null.iloc[:,19:21].head()
```

```python
# Lets check 1st that all the people who had reserved the rooms have got the same exact room type or not 

for g in range(len(df_no_null.loc[:,'reserved_room_type'])):
    if df_no_null.iloc[g,19] != df_no_null.iloc[g,20]:
        print('Some people have not been assigned the same room that they had reserved.')
        break
else:
    print('All the people have been assigned the same room that they had reserved.')
```

Some people have not got the same room type what they reserved for.

```python
# Lets check how many people have not got the expected room.

not_satisfied_customers = []
for q in range(len(df_no_null.loc[:,'reserved_room_type'])):
    if df_no_null.iloc[q,19] != df_no_null.iloc[q,20]:
        not_satisfied_customers.append(q)
len(not_satisfied_customers)
```

```python
# Lets find the %
total_rows = len(df_no_null.loc[:,'reserved_room_type'])

percentage_of_people_who_have_not_got_same_room = (len(not_satisfied_customers) / total_rows) * 100

not_satisfactory_percentage = round(percentage_of_people_who_have_not_got_same_room, 2)

print(f'Percentage of people who did not get the expected room type = {not_satisfactory_percentage} %')
```

```python
# % of people who were satisfied with their booking
no_of_satisfied_customers = 100 - percentage_of_people_who_have_not_got_same_room
no_of_satisfied_customers = round(no_of_satisfied_customers, 2)
print(f'No of statisfied customers with their boking is {no_of_satisfied_customers} %')
```

```python
xindex1 = [0,8]
xindex2 = [3,8]
yindex1 = [no_of_satisfied_customers,0]
yindex2 = [not_satisfactory_percentage,0]
plt.bar(xindex1 ,yindex1 , color = 'green', label = 'Satisfied')
plt.bar(xindex2 ,yindex2 , color = 'red', label = 'Not Satisfied')
# plt.xlabel('Not Satisfied')
plt.ylabel('Percentage of customers')
plt.xticks([0,3],['Satisfied','Not Satisfied'])
plt.legend()
plt.title('Customer satisfaction based on reservation')

for i in range(len(xindex1)) :
    plt.text(xindex1[i],yindex1[i]+1,str(yindex1[i])+ ' %')
    # plt.text(xindex, yindex, the value you want to show)
for i in range(len(xindex2)) :
    plt.text(xindex2[i],yindex2[i]+1,str(yindex2[i])+ ' %')
plt.show()
```
